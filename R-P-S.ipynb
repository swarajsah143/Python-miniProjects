```python
import random

emojis={'s':'‚úÇÔ∏è','p':'üìÉ','r':'ü™®'}
choice=('s','p','r')
    
win=0
lost=0
draw=0

def get_user_choice():
    while True:
        user=input("Rock, Paper,Scissor ? (r/p/s)").lower()
        if user in choice:
            return user
        else:
            print("Invalid!\n")


def display_choices(user,comp):
        print(f"Your choice was {emojis[user]} \nComputer choice was {emojis[comp]}")


def display_result(user,comp):
    if user==comp:
        print("It's a draw!\n")
        global draw
        draw+=1
        print(f"Total draw={draw}")

    elif ((user=='r' and comp=='s') or
          (user=='p' and comp=='r') or  
          (user=='s' and comp=='p')):
        
        print("You won!\n")
        global win
        win+=1

    else:
        print("You lose!\n")
        global lost
        lost+=1
def end_game():
    print("--------Game Over--------\n")
    print(f"Total win={win}\nTotal lost={lost}\nTotal draw={draw}")


def play_game():
    while True:
        user=get_user_choice()
        comp=random.choice(['r','p','s'])
        display_choices(user,comp)
        display_result(user,comp)
    
        continue_playing=input("Do you want to play again? (y/n)\n").lower()
        if continue_playing!='y':
             if continue_playing=='n':
                 end_game()
                 break
             else:
                 continue_playing=input('Invalid! Please press among (y/n)')
                 if continue_playing=='n':
                     end_game()
                     break
                     
        
play_game()

```
